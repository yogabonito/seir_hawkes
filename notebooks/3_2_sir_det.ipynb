# The deterministic SIR model

```python
from py_hawkesn_sir.py_hawkesn_sir.sir_deterministic \
    import DeterministicSIR
```

## Plotting the model with two different sets of parameters

```python
s_0 = 80
i_0 = 20
r_0 = 0
beta1 = 0.5
gamma1 = 0.7
sir1 = DeterministicSIR(s_0=s_0, i_0=i_0, r_0=r_0,
                        beta=beta1, gamma=gamma1)
sir1.solve_euler()
sir1.plot()
```

```python
beta2 = 1.0
gamma2 = 0.3
sir2 = DeterministicSIR(s_0=s_0, i_0=i_0, r_0=r_0,
                        beta=beta2, gamma=gamma2)
sir2.solve_euler()
sir2.plot()
```

## Calculating $S(\infty)$

```python
from math import log

from scipy.optimize import fixed_point

n = s_0 + i_0

def f(x, beta, gamma):
    return i_0 + s_0 + \
           gamma/beta * n * (log(x) - log(s_0))

fixed_point(f, x0=50, args=(beta1, gamma1))
```

```python
fixed_point(f, x0=3, args=(beta2, gamma2))
```
