# The deterministic SEIR model

```python
from py_hawkesn_sir.py_hawkesn_sir.seir_deterministic import DeterministicSEIR
```

## Plotting the model with two different sets of parameters

```python
s_0 = 80
e_0 = 0
i_0 = 20
r_0 = 0
```

```python
seir = DeterministicSEIR(
    s_0=s_0, e_0=e_0, i_0=i_0, r_0=r_0,
    beta=.5, sigma=.3, gamma=.7)
seir.solve_euler()
seir.plot()
```

```python
seir = DeterministicSEIR(
    s_0=s_0, e_0=e_0, i_0=i_0, r_0=r_0,
    beta=1, sigma=.5, gamma=.3)
seir.solve_euler()
seir.plot()
```
